## Visual Analysis



### **Sales Analysis Visualizations**
1. **Maximum Product Sale** – Bar chart showing the product with the highest sales.  
2. **Minimum Product Sale** – Bar chart showing the product with the lowest sales.  
3. **Total Sales** – Line plot of total sales over time.  
4. **Best Product Contribution** – Pie chart showing the contribution of each product to total sales.  
5. **Missing Price Fill with Mean** – Histogram before and after filling missing prices.  
6. **Average Sale Within Given Date Range** – Line plot for sales within a specified period.  
7. **Offer Impact on Best-Selling Product** – Annotated bar chart showing the impact of offers.  
8. **Sales Distribution** – Histogram of sales values.  
9. **Seasonal Sales Trend** – Line plot of sales over months.  
10. **Sales Growth Over Time** – Cumulative line chart showing sales growth.  
11. **Sales by Region** – Grouped bar chart comparing sales across regions.  
12. **Sales Heatmap** – Heatmap for daily/weekly/monthly sales trends.  
13. **Customer Purchase Frequency** – Pie chart or histogram of customer purchases.  
14. **Top 5 Products** – Horizontal bar chart for top-selling products.  
15. **Low-Performing Products** – Bar chart of bottom 5 products.  
16. **Sales Before and After Discount** – Line/bar chart comparing sales before and after discounts.  
17. **Revenue Forecasting** – Line plot showing past and predicted revenue trends.  
18. **Sales vs Advertisement Spending** – Scatter plot with trend line.  
19. **Stock vs Sales Analysis** – Stacked bar chart for stock availability vs sales.  
20. **Product Category Sales** – Grouped bar chart comparing sales across product categories.  




Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
```

Creating a Sample Sales Dataset

```python
# Creating a dataset with product sales across different regions
data = {
    'Product': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E'],
    'Total_Sales': [15000, 12500, 9800, 5600, 20000],
    'Growth_Rate': [12, 8, -5, -12, 15],  # Growth percentage
    'Average_Sale': [500, 450, 300, 150, 600],
    'Region': ['North', 'South', 'East', 'West', 'Metro'],
    'Seasonal_Trend': ['Summer', 'Year-round', 'Winter', 'Festive', 'Holiday'],
    'Stock': [200, 180, 250, 100, 300],
    'Discount_Impact': [20, 10, 5, 0, 25],
    'Ad_Spend': ['Medium', 'High', 'Low', 'Medium', 'High'],
    'Category': ['Electronics', 'Home Goods', 'Apparel', 'Grocery', 'Electronics']
}

df = pd.DataFrame(data)
print(df.head())
```

Data Cleaning - Handling Missing Values

```python
# Simulating missing values
df.loc[2, 'Total_Sales'] = np.nan

# Impute missing values using mean
imputer = SimpleImputer(strategy='mean')
df['Total_Sales'] = imputer.fit_transform(df[['Total_Sales']])

print("After Filling Missing Values:")
print(df)
```

Maximum & Minimum Product Sale (Bar Chart)

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=df['Product'], y=df['Total_Sales'], palette='coolwarm')
plt.title("Product Sales Comparison")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

Total Sales Over Time (Line Plot)

```python
# Simulated monthly sales data for visualization
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
total_sales = [12000, 14000, 15000, 17000, 16500, 18000, 17500, 19000, 18500, 20000, 19500, 21000]

plt.figure(figsize=(10, 5))
plt.plot(months, total_sales, marker='o', linestyle='-', color='b')
plt.title("Total Sales Over Time")
plt.xlabel("Months")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

Best Product Contribution (Pie Chart)

```python
plt.figure(figsize=(7, 7))
plt.pie(df['Total_Sales'], labels=df['Product'], autopct='%1.1f%%', colors=['red', 'blue', 'green', 'orange', 'purple'])
plt.title("Best Product Contribution to Total Sales")
plt.show()
```

Sales Distribution (Histogram)

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Total_Sales'], bins=5, kde=True)
plt.title("Sales Distribution")
plt.xlabel("Total Sales")
plt.ylabel("Frequency")
plt.show()
```

Seasonal Sales Trend (Line Plot)

```python
season_sales = {'Winter': 7000, 'Spring': 12000, 'Summer': 18000, 'Monsoon': 15000, 'Fall': 16000, 'Holiday': 22000}

plt.figure(figsize=(10, 5))
plt.plot(season_sales.keys(), season_sales.values(), marker='o', linestyle='-', color='g')
plt.title("Seasonal Sales Trend")
plt.xlabel("Season")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

Sales vs Advertisement Spending (Scatter Plot)

```python
ad_spend = [10000, 15000, 7000, 12000, 18000]
sales = df['Total_Sales']

plt.figure(figsize=(10, 5))
sns.scatterplot(x=ad_spend, y=sales)
plt.title("Sales vs Advertisement Spending")
plt.xlabel("Advertisement Spend ($)")
plt.ylabel("Total Sales ($)")
plt.grid(True)
plt.show()
```

Stock vs Sales Analysis (Stacked Bar Chart)

```python
plt.figure(figsize=(10, 5))
df.plot(x='Product', y=['Stock', 'Total_Sales'], kind='bar', stacked=True, colormap='viridis', figsize=(10, 5))
plt.title("Stock vs Sales Analysis")
plt.xlabel("Product")
plt.ylabel("Units")
plt.show()
```

Product Category Sales Comparison (Grouped Bar Chart)

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=df['Category'], y=df['Total_Sales'], palette='pastel')
plt.title("Product Category Sales")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

Product Category Sales Comparison (Grouped Bar Chart)

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=df['Category'], y=df['Total_Sales'], palette='pastel')
plt.title("Product Category Sales")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

Revenue Forecasting (Time Series)

```python
# Assuming growth over time
future_months = ['Jan 2026', 'Feb 2026', 'Mar 2026', 'Apr 2026', 'May 2026', 'Jun 2026']
future_sales = [22000, 23000, 24500, 26000, 27500, 29000]

plt.figure(figsize=(10, 5))
plt.plot(months + future_months, total_sales + future_sales, marker='o', linestyle='-', color='r')
plt.title("Revenue Forecasting")
plt.xlabel("Time")
plt.ylabel("Projected Sales")
plt.grid(True)
plt.show()
```
