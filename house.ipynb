```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data
data = {
    'SquareFeet': [1500, 2000, 2500, 1800, 3000, 1200, 2300, 2200, 1700, 1600],
    'Bedrooms': [3, 4, 3, 3, 5, 2, 4, 3, 3, 2],
    'Age': [10, 15, 20, 15, 10, 30, 5, 8, 12, 25],
    'Price': [300000, 450000, 500000, 400000, 600000, 200000, 480000, 470000, 350000, 300000]
}

# Convert to DataFrame
df = pd.DataFrame(data)
```

```python
# Scatter plot
plt.scatter(df['SquareFeet'], df['Price'], color='blue', label='Actual Data')

# Fit the trend line
z = np.polyfit(df['SquareFeet'], df['Price'], 1)
p = np.poly1d(z)
plt.plot(df['SquareFeet'], p(df['SquareFeet']), color='red', label='Trend Line')

# Adding labels and title
plt.xlabel('Square Feet')
plt.ylabel('Price')
plt.title('House Price Trend Line')
plt.legend()

# Show plot
```

Find co-ralation feet vs price

```python
# Calculate the correlation
correlation = df['SquareFeet'].corr(df['Price'])
print(f'Correlation between Square Feet and Price: {correlation}')
```
