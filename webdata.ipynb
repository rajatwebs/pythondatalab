#WEB DATA SCRAPPING USING PANDAS

Beautiful Soup (bs4) is a Python package for parsing HTML and XML documents, including those with malformed markup. It creates a parse tree for documents that can be used to extract data from HTML,[3] which is useful for web scraping.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
url = 'https://www.rbwebsolutions.in/testdata.html'
req = requests.get(url)
print(req)
```

```python
soup = BeautifulSoup(req.content,'html.parser')
tabela = soup.find(name='table')
#print(tabela) #To verify output (VERY LONG)
```

```python
df = pd.read_html(str(tabela))[0].set_index('Order ID') 
#We use "[0]" because the output would be a list with a single element, 
# and we need that element, not the list
df.head()
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Step 2: Send a Request to the Webpage
# Send a request to the webpage to get the HTML content.


url = "https://en.wikipedia.org/wiki/List_of_Nobel_laureates"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
df = pd.read_html(str(soup))
```
